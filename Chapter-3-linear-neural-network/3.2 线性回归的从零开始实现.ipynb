# 3.2 线性回归的从零开始实现

在了解线性回归的关键思想后，我们可以开始通过代码来动手实现线性回归了。在本节中，我们将从零开始实现整个方法，包括数据流水线、模型、损失函数和小批量随机梯度下降优化器。

现代的深度学习框架几乎可以自动化地进行所有这些工作，但从零开始实现可以确保我们真的掌握这些原理，方便我们自定义模型、损失函数等。

## 3.2.1 生成数据集

为简单起见，我们将根据带有噪声的线性模型构造一个数据集。我们的任务是使用这个有限样本的数据集来恢复这个模型的参数。

下面的代码中，我们生成一个包含1000个样本的数据集，每个样本包含从标准正态分布中抽样的两个特征。因此，该数据集可表示为$ \boldsymbol{X} \in \mathbb{R}^{1000 \times 2} $。

我们使用线性模型参数$ \boldsymbol{w} = [2, -3.4]^T $、$ b = 4.2 $和噪声项$ \epsilon $来生成数据集的标签：

$$ \boldsymbol{y} = \boldsymbol{X} \boldsymbol{w} + b + \epsilon $$

其中，$ \epsilon $是假定的服从均值为0，标准差为0.01的高斯噪声。

```python
%matplotlib inline
import torch
from d2l import torch as d2l

def synthetic_data(w, b, num_examples): #@save
    """生成y = Xw + b + 噪声"""
    X = torch.normal(0, 1, (num_examples, len(w)))
    y = torch.matmul(X, w) + b
    y += torch.normal(0, 0.01, y.shape)
    return X, y.reshape(-1, 1)
```

```python
true_w = torch.tensor([2, -3.4])
true_b = torch.tensor(4.2)
features, labels = synthetic_data(true_w, true_b, 1000)

print('features.shape =', features.shape)
print('labels.shape =', labels.shape)
print('features[0] =', features[0])
print('labels[0] =', labels[0])
```

通过生成第二个特征和标签的散点图，可以直观地观察到两者之间的线性关系：

```python
d2l.set_figsize()
d2l.plt.scatter(features[:, (1)].detach().numpy(), labels.detach().numpy(), 1)
```

## 3.2.2 读取数据集

首先，定义一个函数，能从数据集中随机获取小批量样本：

```python
import random
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(indices[i: min(i + batch_size, num_examples)])
        yield features[batch_indices], labels[batch_indices] # 使用yield代替return，可以将函数视作一个迭代器
```

通常，我们利用GPU并行计算的优势，处理大小合适的“小批量”。每个样本都可以被并行地进行模型运算，且每个样本损失函数的梯度也可以被并行计算。GPU可以实现在处理几百个样本时所花费的时间不必处理单个样本时多太多。

```python
batch_size = 16
for i, (X, y) in enumerate(data_iter(batch_size, features, labels)):
    if i == 62:
        print(f'第{i}个小批次数据为：', X, '\n', y) # 共62个小批量
```

上面的迭代会遍历完整个数据集，很直观，但效率很低，在实际工程中并不会这么做。例如，它要求我们将所有数据加载到内存中，并执行大量的随即内存访问。在深度学习框架中，实现了内置的高效迭代器，可以处理存储在文件中的数据和数据流提供的数据。

## 3.2.3 初始化模型参数

在开始用SGD优化模型参数前，需要对参数进行初始化，下面从均值为0，标准差为0.01的正态分布中随机抽样作为权重的初始值，并将偏置初始化为0：

```python
w = torch.normal(0, 0.01, size=(2, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

在初始化后，即可根据损失函数计算梯度，对参数进行更新。由于手动计算梯度很容易出错，因此使用2.5节引入的自动微分来计算梯度。

## 3.2.4 定义模型

定义模型，也就是规定：对于一个给定的输入，通过怎样的计算过程可以获得输出：

```python
def linreg(X, w, b): #@save
    """线性回归模型"""
    '''
    parameters:
    X: shape为(m, n)的二维矩阵，m为样本数，n为特征数
    w: 长度为n的一维向量
    '''
    
    return torch.matmul(X, w) + b
```

## 3.2.5 定义损失函数

参数的梯度是根据损失函数对参数进行求导得到的：

```python
def squared_loss(y_hat, y): #@save
    """均方损失"""
    y.reshape(y_hat.shape)
    loss = ((y_hat - y) ** 2) / 2
    return loss
```

## 3.2.6 定义优化算法

尽管线性回归有解析解，但也可以用梯度下降的方式计算数值解，这里定义小批量随机梯度下降优化算法SGD：

```python
def sgd(params, lr, batch_size): #@save
    """小批量随机梯度下降"""
    '''
    parameters:
    params: 待更新的参数
    lr: learning rate，学习率
    batch_size: 小批量的大小
    '''
    
    with torch.no_grad():
        for param in params:
            param -= lr * param.grad / batch_size # 除以批量的大小，归一化更新步长
            param.grad.zero_()
```

## 3.2.7 训练

至此，模型训练需要的前期准备都已经完成了。

在深度学习实践过程中，会大量重复相似的训练过程：

1. 准备好训练数据集、一组初始化参数、模型、激活函数（这个后面会讲到，它可以看作是模型的一部分）、损失函数、优化算法等必要的工具。
2. 从训练集中抽取小批量样本，并通过模型获得一组预测。
3. 用预测结果和真实标签计算损失函数，并通过反向传播获得参数的梯度。
4. 利用优化算法，对参数进行更新。
5. 重复2~4步骤，直到迭代结束。



在每轮（epoch）中，我们使用data_iter函数遍历整个数据集，并将训练数据集中的所有样本都使用一次。这里的轮数num_epochs和学习率lr都是超参数，分别设为3和0.03。设置超参数是个技术活，非常依赖经验，需要反复试验调整，具体的内容在第11章会讲到。

```python
w = torch.normal(0, 0.01, size=(2, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

def train(num_epochs, lr, loss, model):
    for epoch in range(num_epochs):
        for X, y in data_iter(batch_size, features, labels):
            l = loss(model(X, w, b), y) # 计算小批量中每个样本的损失
            l.mean().backward()
            sgd([w, b], lr, batch_size)

        with torch.no_grad():
            train_l = loss(model(features, w, b), labels)
            print(f'epoch {epoch + 1}, loss {float(train_l.mean()):f}')

lr = 0.1
num_epochs = 15
model = linreg
loss = squared_loss
train(num_epochs, lr, loss, model)
```

由于这个训练集是我们使用已知的模型合成的，因此可以直接将估计的参数与真实的参数进行比较：

```python
print(f'w的估计误差：{true_w - w.reshape(true_w.shape)}')
print(f'b的估计误差：{true_b - b}')
```

事实上，我们几乎不可能完美地求解参数。在机器学习中，我们关心的是如何准确地对参数进行预测而不是获得其真实值，幸运的是，由于深度网络的参数很多，组合丰富，即使是复杂的优化问题，SGD也通常能找到令人满意的解。
